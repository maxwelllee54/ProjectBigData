```python
import pandas as pd

#vol = pd.read_csv('stock_price.csv')
vol = pd.read_csv('avg_price2.csv')
vol['timestamp'] = pd.to_datetime(vol['timestamp'])
vol.fillna(0,inplace=True)
vol.head()
```

```python
#vol.loc[:, 'Cat10'] = vol.iloc[:, 1:11].apply(lambda x: x.argmax(), axis=1)

vol.loc[:, 'before'] = vol.iloc[:, 1:6].mean(axis=1)
vol.loc[:, 'after'] = vol.iloc[:, 5:11].mean(axis=1)

price = vol.copy()

price.loc[vol.before <= 0, 'before'] = 'D'
price.loc[vol.before > 0, 'before'] = 'U'
price.loc[vol.after <= 0, 'after'] = 'D'
price.loc[vol.after > 0, 'after'] = 'U'

price.loc[:, 'Cat'] = price.loc[:, ['before', 'after']].apply(lambda x: ''.join(x), axis=1)

print(price.head())
```
