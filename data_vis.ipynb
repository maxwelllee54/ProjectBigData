```python
import pandas as pd
%matplotlib

df = pd.read_csv('stock_price_240.csv')
df.head()

df2 = pd.read_csv('stock_vol_240.csv')
df2.head()
```

```python
df_ticker = df.groupby('ticker').mean()

df2_ticker = df2.groupby('ticker').mean()
df2_ticker.head()
```

```python
df2_ticker.loc[:, 'Cat10'] = df2_ticker.iloc[:, :].apply(lambda x: x.argmax(), axis=1)
df2_ticker.to_csv('20.csv')

df2_ticker.to_csv('20.csv')
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure()

for ticker in df_ticker.index.values:
    plt.plot(np.arange(-120, 121), df_ticker.loc[ticker, :])

plt.xlim([-120,120])
plt.legend(df_ticker.index.values[:5], loc=3)
    
plt.title('The Ex Ante and Ex Post Stock Return Analysis', fontsize=20)

plt.savefig('stock_240.png')
```
