```python
import pandas as pd

df = pd.read_csv('~/Downloads/DEMO.csv', index_col=None, header=0)
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['release_script'] = pd.to_datetime(df['release_script'])
df.loc[df.timestamp == df.release_script,'close']
```

```python
import numpy as np
def average_price(df):
    
    data = df.copy()
    output = pd.DataFrame(columns=[str(i) for i in np.arange(-120, 121)])
    dic = {}
        
    for date in set(data.loc[:, 'release_script']):

        temp_df1 = data.loc[data.release_script == date, :]

        df_before = temp_df1.loc[temp_df1.timestamp <= temp_df1.release_script, :]

        df_after = temp_df1.loc[temp_df1.timestamp >= temp_df1.release_script, :]
        
        intvl_list = np.arange(120)

        for i in intvl_list:
            try:
                spot_price = (df_before.iloc[-1, 7] + df_after.iloc[0, 7].mean()) /2
                
                avg_before = df_before.iloc[-i:, 7].mean()
                avg_after = df_after.iloc[:i, 7].mean()
            
                output.loc[date, str(-i)] = (spot_price - avg_before) / avg_before
                output.loc[date, str(i)] = (avg_after - spot_price) / spot_price
            except:
                continue
        
    return output

result = average_price(df)
```

```python
import glob
files = glob.glob('final_data/*.csv')

data = pd.DataFrame()

for file in files:
    try:
        ticker = file[file.rfind('_')+1:-4]
        df = pd.read_csv(file, index_col=None, header=0)
        df['timestamp'] = pd.to_datetime(df['timestamp'])
        df['release_script'] = pd.to_datetime(df['release_script'])

        temp_df = average_price(df)
        temp_df.loc[:, 'ticker'] = ticker

        data = pd.concat([data,temp_df])
    except:
        print(file)
```

```python
data.to_csv('avg_price.csv')
```

```python
df_ticker2 = df_new.groupby('ticker').mean()

sensitivity = [str(i) for i in np.arange(-60, 61, 5)]
df_ticker2.loc[:, '0'] = (df_ticker2.loc[:, '-1'] + df_ticker2.loc[:, '1']) /2
```

```python
for i in sensitivity:
    #slope_before = (df_ticker2.loc[:, '0'] - df_ticker2.loc[:, '-'+i]) / float(i)
    
    #slope_after = df_ticker2.loc[:, i] - df_ticker2.loc[:, '0'] / float(i)
    
    #df_ticker2.loc[:, i+'_interval'] = slope_after - slope_before
    if float(i) < 0:
        df_ticker2.loc[:, i+'_mins'] = (df_ticker2.loc[:, '0'] - df_ticker2.loc[:, i]) / float(i)
    
    elif float(i) > 0:
        df_ticker2.loc[:, i+'_mins'] = df_ticker2.loc[:, i] - df_ticker2.loc[:, '0'] / float(i)
    
    else:
        df_ticker2.loc[:, 'annoucement'] = (df_ticker2.loc[:, '1'] - df_ticker2.loc[:, '-1']) / 2
```
