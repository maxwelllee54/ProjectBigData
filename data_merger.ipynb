```python
import pandas as pd

vol = pd.read_csv('stock_vol.csv')
vol['timestamp'] = pd.to_datetime(vol['timestamp'])
vol.fillna(0,inplace=True)
vol.loc[:, 'Cat10'] = vol.iloc[:, 1:11].apply(lambda x: x.argmax(), axis=1)
vol.loc[:, 'Cat2'] = [i[i.find('_')+1:] for i in vol.iloc[:, 1:11].apply(lambda x: x.argmax(), axis=1)]
print(vol.head())
```

```python
sent = pd.read_csv('18.csv')
sent.head()
```

```python
sent["timestamp"] = pd.to_datetime(sent['date'] + ' ' + sent['time'])
```

```python
df = pd.merge(vol, sent, how='left', on=['ticker', 'timestamp'])
df.to_csv('final_result.csv')
```
